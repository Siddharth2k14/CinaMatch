```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/top10K-TMDB-movies.csv', encoding='latin1')
df.head(20)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['tags']=df['overview']+df['genre']
```

```python
new_data=df.drop(columns=['genre','overview'],inplace=True)
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(df['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
df[df['title']=="The Godfather"].index[0]
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
for i in distance[0:5]:
  print(df.iloc[i[0]].title)
```

```python
def recomend(movies):
  index=df[df['title']==movies].index[0]
  distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
  for i in distance[0:5]:
    print(df.iloc[i[0]].title)
```

```python
recomend('Iron Man')
```
